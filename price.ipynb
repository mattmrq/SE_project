```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly as px
import numpy as np

from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("data/PA6_cleaned_dataset.csv")
data['Date'] = pd.to_datetime(data['time'])
data.drop('time', axis=1, inplace=True)

price_odace = pd.read_csv("data/price_odace.csv", sep=";", na_values=["#DIV/0!", "#NA"], decimal=",")
price_odace['Date'] = pd.to_datetime(price_odace['Date'])
price_GV2 = pd.read_csv("data/price_GV2.csv", sep=";", na_values=["#DIV/0!", "#NA"], decimal=",")
price_GV2['Date'] = pd.to_datetime(price_GV2['Date'])
price_IC60 = pd.read_csv("data/price_IC60.csv", sep=";", na_values=["#DIV/0!", "#NA"], decimal=",")
price_IC60['Date'] = pd.to_datetime(price_IC60['Date'])
```

```python
price_odace = price_odace.replace(0, np.nan)
price_GV2 = price_GV2.replace(0, np.nan)
price_IC60 = price_IC60.replace(0, np.nan)

price_odace_filled = price_odace.fillna(method='ffill')
price_GV2_filled = price_GV2.fillna(method='ffill')
price_IC60_filled = price_IC60.fillna(method='ffill')
```

```python
price_odace = price_odace.merge(data[['Date', 'best_price_compound']], on='Date', how='left')

# Merge 'best_price_compound' from 'data' into 'price_GV2'
price_GV2 = price_GV2.merge(data[['Date', 'best_price_compound']], on='Date', how='left')

# Merge 'best_price_compound' from 'data' into 'price_IC60'
price_IC60 = price_IC60.merge(data[['Date', 'best_price_compound']], on='Date', how='left')
```

```python
price_odace.sort_values(by='Date', inplace=True)
price_odace.fillna(method='ffill', inplace=True) 
price_odace.fillna(method='bfill', inplace=True) 
# For price_GV2
price_GV2.sort_values(by='Date', inplace=True)
price_GV2.fillna(method='ffill', inplace=True)
price_GV2.fillna(method='bfill', inplace=True)

# For price_IC60
price_IC60.sort_values(by='Date', inplace=True)
price_IC60.fillna(method='ffill', inplace=True)
price_IC60.fillna(method='bfill', inplace=True)
```

```python
price_odace.set_index('Date', inplace=True)
price_GV2.set_index('Date', inplace=True)
price_IC60.set_index('Date', inplace=True)
```

```python
import plotly.express as px

# Normalizing Function
def normalize_and_plot(df, title):
    scaler = StandardScaler()

    # Normalize the DataFrame
    df_normalized = pd.DataFrame(scaler.fit_transform(df), index=df.index, columns=df.columns)

    # Melt the DataFrame for Plotly
    df_melted = df_normalized.reset_index().melt(id_vars=df_normalized.index.name, var_name='Variable', value_name='Value')

    # Plotting with Plotly
    fig = px.line(df_melted, x=df_normalized.index.name, y='Value', color='Variable', title=title)
    fig.update_layout(width=1300, height=500)
    fig.show()

# Apply to your DataFrames
normalize_and_plot(price_odace, 'All Variables Normalized - price_odace')
normalize_and_plot(price_GV2, 'All Variables Normalized - price_GV2')
normalize_and_plot(price_IC60, 'All Variables Normalized - price_IC60')
```
